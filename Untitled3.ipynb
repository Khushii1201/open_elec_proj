```python
import numpy as np
import skfuzzy as fuzz
from skfuzzy import control as ctrl

# Inputs
dirt = ctrl.Antecedent(np.arange(0, 11, 1), 'dirt')
load = ctrl.Antecedent(np.arange(0, 11, 1), 'load')

# Output
time = ctrl.Consequent(np.arange(0, 61, 1), 'time')

# Membership functions
dirt['low'] = fuzz.trapmf(dirt.universe, [0, 0, 2, 4])
dirt['medium'] = fuzz.trimf(dirt.universe, [2, 5, 7])
dirt['high'] = fuzz.trapmf(dirt.universe, [6, 8, 10, 10])

load['small'] = fuzz.trapmf(load.universe, [0, 0, 2, 4])
load['medium'] = fuzz.trimf(load.universe, [2, 5, 7])
load['large'] = fuzz.trapmf(load.universe, [6, 8, 10, 10])

time['short'] = fuzz.trimf(time.universe, [0, 10, 20])
time['medium'] = fuzz.trimf(time.universe, [15, 30, 45])
time['long'] = fuzz.trimf(time.universe, [40, 50, 60])

# Rules
rule1 = ctrl.Rule(dirt['low'] & load['small'], time['short'])
rule2 = ctrl.Rule(dirt['low'] & load['medium'], time['medium'])
rule3 = ctrl.Rule(dirt['medium'] & load['small'], time['medium'])
rule4 = ctrl.Rule(dirt['medium'] & load['large'], time['long'])
rule5 = ctrl.Rule(dirt['high'] & load['medium'], time['long'])
rule6 = ctrl.Rule(dirt['high'] & load['large'], time['long'])

# System
system = ctrl.ControlSystem([rule1, rule2, rule3, rule4, rule5, rule6])
sim = ctrl.ControlSystemSimulation(system)

# Input values
sim.input['dirt'] = 7
sim.input['load'] = 6

# Compute
sim.compute()

print("Cycle Time:", sim.output['time'])
time.view(sim=sim)
```
